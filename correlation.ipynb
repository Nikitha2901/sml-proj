```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

# Import file with data
data = pd.read_csv("training_data.csv")
data = data.drop(['snow'],axis=1)
data['increase_stock'] = data['increase_stock'].apply(lambda x: 0 if x == 'low_bike_demand' else 1)

# Print correlation matrix
corr_matrix = data.corr()
print(corr_matrix)

# Plotting correlation heatmap
plt.figure(figsize=(12, 10))  # Adjust the figure size as needed
dataplot = sb.heatmap(corr_matrix, cmap="YlGnBu", annot=True, fmt=".2f", annot_kws={"size": 10})

# Add title for better context
plt.title("Correlation Heatmap", fontsize=16)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display heatmap
plt.tight_layout()
plt.show()
```
